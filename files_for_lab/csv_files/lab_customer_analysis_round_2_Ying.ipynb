```python
def clean_data(filename):
    data = pd.read_csv(filename)
    data = data.drop(data.columns[0], axis=1)
    data_cols = [col_name.lower().replace(' ','_') for col_name in data.columns]
    data.columns = data_cols
    data = data.drop(columns='vehicle_type')
    data = data.dropna(subset=['state','response','months_since_last_claim','number_of_open_complaints','vehicle_class','vehicle_size'])
    data['effective_to_month'] = pd.to_datetime(data['effective_to_date'], errors='coerce').dt.month
    cleaned_data = data[data['effective_to_month'] < 4]
    return cleaned_data
```

```python
clean_data("marketing_customer_analysis.csv")
```
